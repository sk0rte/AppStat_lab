```python
name = np.array(["sejr", "chamilla", "daniel", "emil", "julius"])

def fit(x, a, b):
    return a * x + b

def gauss_pdf(x, mu, sigma) :
    global binwidth, N_points
    """Gaussian"""
    return binwidth * N_points * 1.0 / np.sqrt(2*np.pi) / sigma * np.exp( -0.5 * (x-mu)**2 / sigma**2)
```

```python
for i in name:
    xmin = -0.5
    xmax = 0.5
    Data = np.genfromtxt(i + "_usikkerhed.dat")
    if i == "emil":
        Data = Data[:-3]
    n = Data[:,0]
    tid = Data[:,1]
    fig, ax = plt.subplots(figsize=(16, 6), nrows=2, ncols=2)
    ax[0,0].plot(n, tid, ".", label="Data")
    ax[0,0].set_title(i)
    ax[0,0].set_xlabel("nuber measurment")
    ax[0,0].set_ylabel("time")
    par = np.polyfit(n, tid, 1)
    Y_lin = fit(n, *par)
    res = tid - fit(n, *par)
    ax[0,0].plot(n, Y_lin, label="Fit")
    ax[1,0].plot(n, res, ".")
    ax[0,1].hist(res, bins=20, range=(-0.5, 0.5), histtype="step")
    N_points = len(res)
    binwidth = (xmax - xmin)/ N_points
    ullhfit = UnbinnedLH(gauss_pdf, res, bound=(xmin, xmax), extended=True)
    minuit_ullh = Minuit(ullhfit, mu=0, sigma=0.1)
    minuit_ullh.errordef = 0.5 
    minuit_ullh.migrad();
    plt.show()
    print(i, minuit_ullh.values[:][1], "s", "minuits unbinned likelyhoodfit")
    print(i, np.std(res), "s", "ren standard afigelse")

    
    
    
    
    
    
    
    
    #print(i, np.std(Data_dif))
```
